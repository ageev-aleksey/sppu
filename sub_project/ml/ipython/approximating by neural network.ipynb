```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow.keras as k
```

```python
def foo(x):
    return x*np.sin(0.5*x)*np.cos(x/0.5)
```

```python
x = np.linspace(-10, 10, 500)
y = foo(x)
plt.plot(x, y)
```

```python
inputs = k.Input(shape=(1,))
layer = k.layers.Dense(32, activation='sigmoid')(inputs)
layer = k.layers.Dense(32, activation='softplus')(layer)
outputs = k.layers.Dense(1, name ='contorl')(layer)
model = k.Model(inputs=inputs, outputs=outputs , name='test')
model.summary()
```

```python
model.compile(loss='MSE', optimizer='RMSprop', metrics=['MAE'])
```

```python
model.fit(x.reshape(-1, 1), y.reshape(-1, 1), batch_size=64, epochs=20000)
```

```python
yp = model.predict(x.reshape(-1, 1))
y = foo(x)
plt.plot(x, yp.reshape(-1))
plt.plot(x, y)
```
